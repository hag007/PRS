```bash
python -c '
import pandas as pd
import numpy as np 

def get_summary_row(fname, colnames):

    d_cols={
        "bca_313_neg_cimba_eur_brca1_oncoarray": "BRCA1 EUR ER-neg",
        "bca_313_neg_cimba_aj_brca1_oncoarray": "BRCA1 AJ ER-neg",
        "bca_313_cimba_eur_brca1_oncoarray": "BRCA1 EUR Overall",
        "bca_313_cimba_aj_brca1_oncoarray": "BRCA1 AJ Overall",
        "bca_313_cimba_eur_brca2_oncoarray": "BRCA2 EUR Overall",
        "bca_313_cimba_aj_brca2_oncoarray": "BRCA2 AJ Overall",

        "bcac_onco_eur-er-neg-5pcs_cimba_eur_brca1_oncoarray": "BRCA1 EUR ER-neg",
        "bcac_onco_eur-er-neg-5pcs_cimba_aj_brca1_oncoarray": "BRCA1 AJ ER-neg",
        "bcac_onco_eur-5pcs_cimba_eur_brca1_oncoarray": "BRCA1 EUR Overall",
        "bcac_onco_eur-5pcs_cimba_aj_brca1_oncoarray": "BRCA1 AJ Overall",
        "bcac_onco_eur-5pcs_cimba_eur_brca2_oncoarray": "BRCA2 EUR Overall",
        "bcac_onco_eur-5pcs_cimba_aj_brca2_oncoarray": "BRCA2 AJ Overall"
    }
    
    d_methods={
        "brca_313_metrics": "313-PRS",
        "brca_pt3_metrics": "P+T",
        "brca_ls_metrics": "Lassosum"
    }

    df=pd.read_csv("'$PRS_OUTPUT'/"+fname+".tsv", sep="\t")
    idx=df.apply(lambda a: d_cols[a["discovery"]+"_"+a["target"]], axis=1)
    idx.name="name"
    df_hr=pd.concat((idx, df[["hr.exp.beta"]+colnames]), axis=1)
    df_hr_gb=df_hr.groupby("name")
    serieses=[]
    for cur_i, cur_df in df_hr_gb:
        serieses.append(cur_df.loc[cur_df.iloc[:,1].idxmax()])
    df_hr=pd.concat(serieses,axis=1).T
    df_hr.iloc[:,1:]=df_hr.iloc[:,1:].astype(float)
    df_hr.index=df_hr.iloc[:,0]
    df_hr_sum_col=df_hr.apply(lambda a : f"{round(a.iloc[2],3)} ({round(a.iloc[3],3)}, {round(a.iloc[4],3)})", axis=1)
    df_hr_sum_col.columns=df_hr.columns
    df_hr_sum_col.name=d_methods[fname]
    return df_hr_sum_col


df_hr=pd.concat((get_summary_row("brca_313_metrics", ["hr.exp.beta","hr.ci.min","hr.ci.max"]),
                get_summary_row("brca_pt3_metrics", ["hr.exp.beta","hr.ci.min","hr.ci.max"]),
                get_summary_row("brca_ls_metrics", ["hr.exp.beta","hr.ci.min","hr.ci.max"])), axis=1).T
df_hr.to_csv("'$PRS_OUTPUT'/BRCA_HR.tsv", sep="\t")

df_hr10=pd.concat((get_summary_row("brca_313_metrics", ["hr.10","hr.10.ci.min","hr.10.ci.max"]),
            get_summary_row("brca_pt3_metrics", ["hr.10","hr.10.ci.min","hr.10.ci.max"]),
            get_summary_row("brca_ls_metrics", ["hr.10","hr.10.ci.min","hr.10.ci.max"])), axis=1).T
df_hr10.to_csv("'$PRS_OUTPUT'/BRCA_HR10.tsv", sep="\t")

df_hr90=pd.concat((get_summary_row("brca_313_metrics", ["hr.90","hr.90.ci.min","hr.90.ci.max"]),
            get_summary_row("brca_pt3_metrics", ["hr.90","hr.90.ci.min","hr.90.ci.max"]),
            get_summary_row("brca_ls_metrics", ["hr.90","hr.90.ci.min","hr.90.ci.max"])), axis=1).T
df_hr90.to_csv("'$PRS_OUTPUT'/BRCA_HR90.tsv", sep="\t")
'
```
